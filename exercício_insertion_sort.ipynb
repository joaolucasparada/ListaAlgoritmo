```python
import time
import random
import matplotlib.pyplot as plt
import pandas as pd

def insertion_sort(lista):
    n=len(lista)
    tempo_inicial= time.time()
    for p in range(1):
        for i in range(1,n):
            temporario= lista[i]
            j= i-1
            while j >=0 and  lista[j] > temporario:
                lista[j+1]= lista[j]
                j-=1
            lista[j+1]=temporario
    tempo_final= time.time()
    tempo_total= tempo_final - tempo_inicial
    return  tempo_total, lista
```

```python
tempo_de_execução= {'Tamanho_da_lista':[],
                    'lista_ordenada':[],
                    'lista_aleatória':[],
                    'lista_invertida':[]
                    
                    }

for i in range(50,101,50):
    list_ordenada = [x for x in range(i)]
    list_aleatória= [random.randint(1, 500) for _ in range(i)]
    list_invertida= [x for x in range(i, -1,-1)]
    temp_list_ordenada = insertion_sort(list_ordenada)[0]
    temp_list_invertida= insertion_sort(list_invertida)[0]
    temp_list_aleatória= insertion_sort(list_aleatória)[0]   
    

    tempo_de_execução['Tamanho_da_lista'].append(i)
    tempo_de_execução['lista_invertida'].append(temp_list_invertida)
    tempo_de_execução['lista_ordenada'].append(temp_list_ordenada)
    tempo_de_execução['lista_aleatória'].append(temp_list_aleatória)
```

```python
# Criando o gráfico
plt.figure(figsize=(10, 6))

plt.plot(tempo_de_execução['Tamanho_da_lista'], tempo_de_execução['lista_invertida'], label="Invertida", marker='o', color='b')
plt.plot(tempo_de_execução['Tamanho_da_lista'], tempo_de_execução['lista_ordenada'], label="Ordenada", marker='s', color='g')
plt.plot(tempo_de_execução['Tamanho_da_lista'], tempo_de_execução['lista_aleatória'], label="Aleatória", marker='^', color='r')

plt.xlabel("Tamanho da lista")
plt.ylabel("Tempo de execução (segundos)")
plt.title("Tempo de execução em diferentes tipos de listas")

plt.legend()
plt.grid(True)

plt.show()
```
