```python
import time
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def busca_binaria(alvo, arr, arr_fields, tamanho):
        inferior = 0
        superior = tamanho

        while(inferior <= superior):
            mid =  (inferior + superior) // 2

            if mid < tamanho:
                if arr[mid] == alvo:
                    return arr[mid], arr_fields[mid]
                elif arr[mid] < alvo:
                        inferior = mid + 1
                elif arr[mid] > alvo:
                    superior = mid - 1
            else:
                return print("Chave n√£o encontrada")

def cronometrar_busca_binaria(alvo, arr, arr_fields, tamanho, lacos=1000000):
    inicio = time.time()
    for p in range(lacos):
        resultado = busca_binaria(alvo, arr, arr_fields, tamanho)
    fim = time.time()
    tempo_total = fim - inicio
    return resultado, tempo_total
```

```python
L = [k for k in range(0, 11)]
n = len(L)
Letras = ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J", "K"]
```

```python
y = []
for i in range(0,11):
    y.append(cronometrar_busca_binaria(alvo=i, arr=L, arr_fields=Letras, tamanho=n, lacos=10000))

df = pd.DataFrame(y, columns=['Campo', 'Tempo'])
print(df)
```

```python
y = []
for i in range(0,11):
    y.append(cronometrar_busca_binaria(alvo=i, arr=L, arr_fields=Letras, tamanho=n)[1])
print(y)
```
